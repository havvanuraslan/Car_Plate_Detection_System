```python
from ultralytics import YOLO

model = YOLO('yolov8x.pt')

results = model.train(
    data="/content/drive/MyDrive/dataset/new_data.yaml",
    epochs=20,
    imgsz=640,
    batch=16,
    workers=8,
    amp=True,
    cache=True,
    optimizer='SGD',
    lr0=0.001,
    lrf=0.01,
    weight_decay=0.0005,
    augment=True,
    patience=50,
    freeze=10,
    save=True,
    plots=True
)
```

```python
import os
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO


model = YOLO('/content/runs/detect/train/weights/best.pt')

img_path = '/content/qqAAAgCe0OA-960.jpg'

results = model.predict(source=img_path, conf=0.5, save=True)

pred_img_name = os.path.basename(img_path)
pred_img_path = f'runs/detect/predict/{pred_img_name}'

pred_img = cv2.imread(pred_img_path)
pred_img = cv2.cvtColor(pred_img, cv2.COLOR_BGR2RGB)
plt.imshow(pred_img)
plt.axis('off')
plt.show()
```
